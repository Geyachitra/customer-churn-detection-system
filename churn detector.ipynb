```python
Y=df.Churn
Y
```

```python
X=df[df.columns.difference(['State','Area Code','Phone','Churn'])]
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
logistic_model=log_reg.fit(X_train,y_train)
logistic_model
```

```python
predictions = logistic_model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_test,predictions)
```

```python
X_new=[[121,3,88,37.09,218.2,108,29.62,348.5,7,2.03,7.5,0,118,9.57,212.6,24,1]]
y_new=log_reg.predict(X_new)
print("X=%s, Predicted=%s" % (X_new[0], y_new[0]))
```

```python
from sklearn import tree
decisiontree_model = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
dtree_fit = decisiontree_model.fit(X_train,y_train)
```

```python
predictions = dtree_fit.predict(X_test)
```

```python
metrics.confusion_matrix(y_test,predictions)
```
